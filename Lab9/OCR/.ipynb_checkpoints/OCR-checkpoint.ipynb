```python
def draw_image(img, figsize=(6, 6)):
    """draws the image"""
    fig, ax = plt.subplots(figsize=figsize)
    ax.imshow(img, cmap='Greys')
    plt.show()


def plot_both(img, corr, name='pattern', figsize=(6, 6)):
    """draws the given image and it's correlation as points"""
    fig, ax = plt.subplots(figsize=figsize)
    ax.imshow(img, cmap='Greys')
    points = list(zip(
        *[(x, y) for y, row in enumerate(corr) for x, v in enumerate(row) if
          v != 0]))
    ax.plot(points[0], points[1], 'ro', ms=2.0, label=name)
    ax.legend()
    plt.show()
```

```python
def get_OCR_efficiency(text_predefined, text_read):
    """returns the OCR efficiency using levenshteinDistance"""
    text_read = text_read.replace("\n", " ")
    return 1 - float(levenshteinDistance(text_predefined, text_read) - 5) / len(text_predefined)    
```

```python
def levenshteinDistance(s1, s2):
    """returns minimal number of operations [swap, remove, add] to similarize texts"""
    if len(s1) > len(s2):
        s1, s2 = s2, s1

    distances = range(len(s1) + 1)
    for i2, c2 in enumerate(s2):
        distances_ = [i2+1]
        for i1, c1 in enumerate(s1):
            if c1 == c2:
                distances_.append(distances[i1])
            else:
                distances_.append(1 + min((distances[i1], distances[i1 + 1], distances_[-1])))
        distances = distances_
    return distances[-1]
```

```python
def get_number_of_occurencies(text_read):
    """returns map with number of occurencies for every letter"""
    result = dict()
    for letter in font:
        result[letter] = len(list(filter(lambda x: x == letter, text_read)))
    return result
```

```python
def get_sorted_alphabet():
    """returns alphabet + numbers in order stored in the alphabet image"""
    letters_sorted = sorted(list(string.ascii_lowercase)) + [str(num) for num in range(1, 10)] + ['0']
    if include_coma_and_dot:
        letters_sorted = letters_sorted + [',', '.']
    letters_sorted += ['?', '!']
    return letters_sorted
```

# Znajdź możliwe wystąpienia dla każdej z liter:

```python
def get_occurences_of_letters(img, font, threshold=0.7):
    """try to fit letters in image"""
    corrs = dict()
    fi = np.fft.fft2(img)
    for letter in tqdm(font, desc="reading correlations"):
        fp = np.fft.fft2(np.rot90(font[letter], 2), fi.shape)  # nie pasuja wymiary to dopisze automagicznie zera
        m = np.multiply(fi, fp)
        corr = np.fft.ifft2(m)
        corr = np.abs(corr)
        corr = corr.astype(float)
        corr[corr < threshold * np.amax(corr)] = 0
        corrs[letter] = corr
    return corrs
```

# Funkcje do czytania plików

```python
def load_font():
    """returns a font dictionary"""
    font = dict()
    font_dir = directory
    font_dir += 'glyph/' + font_type

    for filename in os.listdir(font_dir):
        if len(filename) == 5:
            font[filename[0]] = 255 - ndimage.imread(font_dir + '/' + filename, flatten=True)
        elif filename == 'dot.png' and include_coma_and_dot:
            font['.'] = 255 - ndimage.imread(font_dir + '/' + filename, flatten=True)
    if not include_coma_and_dot:
        del font[',']
    return font
```

```python
def read_image(filename):
    """reads image and inverts colors"""
    return 255 - ndimage.imread(directory + filename, flatten=True)

def read_alphabet():
    img_filename = '{0}/alphabet_{1}{2}.png'.format(font_type, font_type, '' if include_coma_and_dot else '_comaless')
    return read_image(img_filename)

def add_alphabet_line(img):
    """adds a line with all letters to the image"""
    alph = read_alphabet()
    conc = np.pad(alph, ((0, 0), (0, img.shape[1] - alph.shape[1])), 'constant', constant_values=0)
    return np.append(conc, img).reshape(conc.shape[0] + img.shape[0], img.shape[1])

def read_facebook(rotation=0):
    return add_alphabet_line(read_image(
        'facebook' + font_type + 'punctation_rot{0}.png'.format(rotation)))

def read_justice():
    return add_alphabet_line(
        read_image('justice' + font_type + 'punctation.png'))
```

## Sprawdzenie poprawności rozpoznania:

Aby sprawdzić jakie litery zostały rozpoznane przy wyszukiwaniu danej litery muszę wiedzieć jaka litera w którym miejscu alfabetu się znajduje. Rozpoznaję to poprzez odnalezienie wszystkich spacji w alfabecie:

```python
def get_all_empty_columns(img):
    """returns empty columns indexes"""
    spaces = []
    for i, col in enumerate(np.transpose(img)):
        if not col.any():
            spaces.append(i)
    return spaces
```

```python
def aggregate_empty_columns_to_spaces(spaces):
    """gets the middle index from clusters of empty columns """
    first = spaces[0]
    last = spaces[0] - 1
    aggregated_spaces = []
    for space in spaces:
        if space == last + 1:
            last = space
        else:
            aggregated_spaces.append((last + first) // 2)
            first = space
            last = space
    aggregated_spaces.append((last + first) // 2)

    return aggregated_spaces[1:]
```

```python
spaces = get_all_empty_columns(img)
aggregated_spaces = aggregate_empty_columns_to_spaces(spaces)
```

Funkcja get_letters_end zwraca słownik. Słownik dla każdej litery posiada kolumnę, w której litera na obrazku się kończy (plus połowa długości spacji)

```python
def get_letters_end(aggregated_spaces):
    letters_sorted = get_sorted_alphabet()
    letter_end = dict()

    for i, pos in enumerate(aggregated_spaces):
        letter_end[letters_sorted[i]] = pos
    if not high_res:
        letter_end['!'] = 633

    return letter_end
```

```python
letter_end = get_letters_end(aggregated_spaces)
```

```python
def get_letter_by_point(point, letter_end):
    "returns letter for given letters end and point, that has been recognised"
    for letter in get_sorted_alphabet():
        if point > letter_end[letter]:
            continue
        else:
            return letter
```

```python
def get_recognised_letters_for_each_letter(corrs, letters_end):
    """Returns dictionary, that for each letters returns letters that have possibly 
    been recognised"""
    letters_found = dict()

    for letter in get_sorted_alphabet():
        letters_found[letter] = set()
        points = list(zip(*[(x, y) for y, row in enumerate(corrs[letter]) for x, v in enumerate(row) if v != 0]))
        points[0] = sorted(points[0])

        for point in points[0]:
            letters_found[letter].add(get_letter_by_point(point, letters_end))

    return letters_found
```

```python
letters_recognised = get_recognised_letters_for_each_letter(corrs, letter_end)
```

## Graf rozpoznań liter

Aby wiedzieć dokładnie które litery są rozpoznawane jako część innych i posortować je topologicznie tworzę graf

```python
def create_graph_from_letters_matches(letters_found):
    edges = [(letter, matched_letter) for letter, matched_letters in letters_found.items() for matched_letter in matched_letters if matched_letter != letter]
    G = nx.DiGraph(edges)
    G.add_nodes_from(list(letters_found.keys()))
    return G
```

```python
topological_alphabet = list(reversed(list(nx.topological_sort(G))))
print(topological_alphabet)
```

```python
def get_topological_alphabet():
    """returns topologically sorted alphabet for globally parametrised font"""
    img = read_alphabet()
    font = load_font()

    corrs = get_occurences_of_letters(img, font, threshold=0.91)

    aggregated_spaces = aggregate_empty_columns_to_spaces(get_all_empty_columns(img))

    letter_end = get_letters_end(aggregated_spaces)

    letters_found = get_recognised_letters_for_each_letter(corrs, letter_end,)

    G = create_graph_from_letters_matches(letters_found)

    return list(reversed(list(nx.topological_sort(G))))
```

### Posortuj dopasowania liter znormalizowane względem najlepszego dopasowania

```python
def get_matchings(corrs, topological_alphabet):
    """get matchings from correlations and returns sorted by matching"""
    matches = []
    for fit_by_letter in tqdm(corrs.items(), desc='processing matchings from correlations'):
        fitted_letter = fit_by_letter[0]
        max_level_fit = np.amax(fit_by_letter[1])
        letter_fitted_ind = np.nonzero(corrs[fitted_letter])
        for i, row in enumerate(letter_fitted_ind[0]):
            col = letter_fitted_ind[1][i]
            match_level = fit_by_letter[1][row, col] / max_level_fit + (0.05 / (1 + topological_alphabet.index(fitted_letter)))
            matches.append((row, col, match_level, fitted_letter))
    return sorted(matches, key=lambda x: x[2], reverse=True)
```

```python
def get_whole_text(text, font):
    """prepare text from filtered list of matchings"""
    tmp_ft = text.copy()
    tmp_ft = [(row, col, c) for row, col, c in tmp_ft if row > min(tmp_ft, key=lambda x: x[0])[0] + (0.55 * font['a'].shape[0]) + 1]
    whole_str = ""
    while len(tmp_ft) > 0:
        line1 = [(row, col, c) for row, col, c in tmp_ft if row < min(tmp_ft, key=lambda x: x[0])[0] + (0.55 * font['a'].shape[0]) + 1]
        line1.sort(key=lambda x: x[1])
        line_str = ""
        at_x = line1[0][1]
        for row, col, c in line1:
            if col - len(font[c][0]) > at_x + (0.5 * font['a'].shape[0]):
                line_str += ' '
            line_str += c
            at_x = col
        whole_str += line_str + "\n"
        for c in line1:
            tmp_ft.remove(c)
            
    return whole_str
```

```python
def create_text_from_matchings(img, matches, font):
    """Creates a full text from all matchings"""

    def match_letter_taken(taken_map, match):
        """Match field of the letter as taken on taken_map"""
        move = 1
        for i in range(match[0] - font[match[3]].shape[0] + move, match[0] + move):
            for j in range(match[1] - font[match[3]].shape[1] + move, match[1] + move):
                taken_map[i, j] = 1

    def letter_overlaps(taken_map, match, font):
        """Check if any of the fields that the match would cover is """
        move = 1
        for i in range(match[0] - font[match[3]].shape[0] + move, match[0] + move):
            for j in range(match[1] - font[match[3]].shape[1] + move, match[1] + move):
                if taken_map[i, j] != 0:
                    return True
        return False

    taken_map = np.zeros(img.shape)
    text = []
    for match in tqdm(matches, desc='filtering matchings'):
        if not letter_overlaps(taken_map, match, font):
            match_letter_taken(taken_map, match)
            text.append(match)

    return get_whole_text(list(map(lambda x: (x[0], x[1], x[3]), text)), font)
```

# Rotacja

```python
def binary_image_with_lines(img):
    """Change lines of text into blocks."""
    tim = ndimage.grey_closing(img, (10, 10))
    tim = tim > np.amax(tim)*0.5
    return tim
```

```python
def draw_approx_fitting(x,y,m,c):
    """plot linear function approximated from x and y"""
    plt.plot(x, y, 'o', label='Original data', markersize=10)
    plt.plot(x, m*x + c, 'r', label='Fitted line')
    plt.legend()
    plt.show()
```

```python
def rgb2gray(rgb):
    """change rgb image to grayscale"""
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])    
```

```python
def load_image_with_rotation(filename):
    """load image and correct rotation using dft and approximation"""
    img = read_image(filename)
    return my_rotate_image(img)

def my_rotate_image(img):
    """rotate image using dft and approximation"""
    fi = np.fft.fft2(binary_image_with_lines(img))

    preprocess_f = np.abs(np.fft.fftshift(fi))
    preprocess_f[preprocess_f < 0.1 * np.amax(preprocess_f)] = 0
    non_zer = np.nonzero(preprocess_f)

    plot_both(img, preprocess_f)

    x = non_zer[1]
    y = -non_zer[0]
    A = np.vstack([x, np.ones(len(x))]).T

    tg_alf, c = np.linalg.lstsq(A, y)[0]
    draw_approx_fitting(x,y, tg_alf, c)
    angle = 90 - np.rad2deg(math.atan(tg_alf))

    return ndimage.rotate(input=img, angle=angle)
```

```python
img = load_image_with_rotation(font_type + '/facebook'+ font_type + 'punctation_rot1.png')
draw_image(img)
```

```python
def load_image_with_correct_rotation_opencv(filename):
    """load image and correct rotation using c"""
    image = cv2.imread(filename)    
    return rotate_image(image)
    
def rotate_image(image):
    """rotate image using cv2"""
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    gray = cv2.bitwise_not(gray)
    
    thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]
    
    coords = np.column_stack(np.where(thresh > 0))
    angle = cv2.minAreaRect(coords)[-1]
    
    if angle < -45:
        angle = -(90 + angle)
    else:
        angle = -angle
    
    (h, w) = image.shape[:2]
    center = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D(center, angle, 1.0)
    rotated = cv2.warpAffine(image, M, (w, h), flags=cv2.INTER_CUBIC, borderMode=cv2.BORDER_REPLICATE)

    return rotated
```

# Efekt końcowy

```python
def ocr(filename):
    img = load_image_with_correct_rotation_opencv(filename)  # Obracanie
    img = cv2.fastNlMeansDenoisingColored(img, None, 10, 10, 7, 21)  # Odszumianie
    img = add_alphabet_line(255 - rgb2gray(img))
    font = load_font()
    topological_alphabet = get_topological_alphabet()
    corrs = get_occurences_of_letters(img, font)
    text = create_text_from_matchings(img, get_matchings(corrs, topological_alphabet), font)
    return corrs, text
```
